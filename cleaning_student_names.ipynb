```python
import pandas as pd
import os
```

```python
file_to_read = os.path.join('Resources','students_complete.csv')
```

```python
students_df = pd.read_csv(file_to_read)
```

```python
students_df.head()
```

```python
student_names = students_df.student_name.tolist()
```

```python
students_df.head()
```

```python
students_to_fix = []
```

```python
for name in student_names:
    if len(name.split()) >= 3:
        students_to_fix.append(name)
len(students_to_fix)
```

```python
prefixes = []
for name in students_to_fix:
    if len(name.split()[0]) <= 4 and len(name.split()) >= 3:
        prefixes.append(name.split()[0])
prefixes = set(prefixes)
prefixes
```

```python
suffixes = []
for name in students_to_fix:
    if len(name.split()[-1]) <= 3 and name.split()[-1] not in suffixes and len(name.split()) >= 3:
        suffixes.append(name.split()[-1])
suffixes = set(suffixes)
suffixes
```

```python
fixed_names = []
pre_and_suff = ['Dr. ','Mrs. ','Mr. ','Ms. ','Miss ',' DDS',' MD',' DVM',' PhD']
for name in students_to_fix:
    new_name = name
    for pre_suff in pre_and_suff:
        new_name = new_name.replace(pre_suff,'')
    fixed_names.append(new_name)
fixed_names
```
